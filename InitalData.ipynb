Most Popular Items

```python
#Purchase count
purchase_count = purchase_df.groupby(["Item ID","Item Name"]).count()["Purchase ID"]
#item price
item_price = purchase_df.groupby(["Item ID","Item Name"]).mean()["Price"]
#Total Purchase Value 
total_pur_value = purchase_count * item_price
#purchase_df.groupby(["Item ID","Item Name"]).sum()["Price"]

popular_df = pd.DataFrame ({     "Purchase Count":purchase_count,
                                 "Item Price":item_price,
                                 "Total Purchase Value":total_pur_value
                           })

popular_df.style.format({"Purchase Count" : "{:,.2f}", 
                        "Item Price": "${:,.2f}", 
                         "Total Purchase Value": "${:,.2f}"}) 
popular_df.sort_values("Purchase Count", ascending = False).head(5)
```

Most Profitable 5 Items

```python

purchase_count = purchase_df.groupby(["Item ID","Item Name"]).count()["Price"]
#item price
item_price = purchase_df.groupby(["Item ID","Item Name"]).mean()["Price"]
#Total Purchase Value 
total_pur_value = purchase_df.groupby(["Item ID","Item Name"]).sum()["Price"]

profitable_df = pd.DataFrame ({
                                 "Purchase Count": purchase_count,
                                 "Item Price": item_price,
                                 "Total Purchase Value":total_pur_value
                           })
profitable_df.style.format({"Purchase Count" : "{:,.2f}", 
                              "Item Price": "${:,.2f}", 
                             "Total Purchase Value": "${:,.2f}"}) 
profitable_df.sort_values("Total Purchase Value", ascending = False).head(5)
```
